# Creating the function

```python
def define_bonus(month,name):
    table = month[["C.I.N","APELLIDOS_NOMBRES","TOTAL HABERES"]]
    all_months = 12
    table["AGUINALDO"] = table["TOTAL HABERES"] / all_months
    print(table.head(60))
    print(table["AGUINALDO"].sum())
    with pd.ExcelWriter(f"{name}_aguinaldo.xlsx") as writer:
        table.to_excel(writer,sheet_name=f'{name}')
```
